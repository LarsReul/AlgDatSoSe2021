```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
def plot_function(_x_min, _x_max, _fl):
    _x = np.linspace(_x_min,_x_max,100)
    cmap = plt.get_cmap("tab10")
    fig = plt.gcf()
    fig.set_size_inches(12.5, 12.5, forward=True)
    for k in range(len(_fl)):
        _y = _fl[k](_x)[0]
        plt.plot(_x,_y,c=cmap(k), label = 'y = '+ _fl[k](1)[1])
        plt.xlabel('x')
        plt.ylabel('y')
    plt.legend(loc='upper left')
```

```python
def fun1(x):
    return np.log2(x), 'lg(x)'
def fun2(x):
    return np.sqrt(x), 'sqrt(x)'
def fun3(x):
    return x, 'x'
def fun4(x):
    return np.log2(x)*x, 'x*lgx'
def fun5(x):
    return x**(3/2), 'x^(3/2)'
def fun6(x):
    return x**2, 'x^2'
def fun7(x):
    return x**3, 'x^3'
def fun8(x):
    return 2**x, '2^x'

fl = [fun1, fun2, fun3, fun4, fun5, fun6, fun7, fun8]
```

```python
indices = [0,1,2,3,4,5,6,7]
plot_function(1,12,[fl[i] for i in indices])
```
